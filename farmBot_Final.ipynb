```python
import torch
import numpy as np
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM

# Step 1: Hindi Speech-to-Text using Hugging Face ASR model
def transcribe_hindi(audio):
    transcriber = pipeline(
        "automatic-speech-recognition",
        model="theainerd/Wav2Vec2-large-xlsr-hindi"
    )
    sr, y = audio
    y = y.astype(np.float32)
    y /= np.max(np.abs(y))
    result = transcriber({"sampling_rate": sr, "raw": y})
    return result["text"]


# Step 2: Hindi → English Translation using NLLB-200
trans_model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
trans_tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")

def translate_to_english(hindi_text):
    translator = pipeline(
        "translation",
        model=trans_model,
        tokenizer=trans_tokenizer,
        src_lang="hin_Deva",
        tgt_lang="eng_Latn",
        max_length=400
    )
    result = translator(hindi_text)
    return result[0]['translation_text']
```

```python
import librosa

# Load Hindi audio (must be 16kHz mono)
audio_path = "/content/hindi_output.mp3"
y, sr = librosa.load(audio_path, sr=16000)
audio = (sr, y)

# Run pipeline
hindi_text = transcribe_hindi(audio)
print("Hindi Text:", hindi_text)

english_text = translate_to_english(hindi_text)
print("English Translation:", english_text)
```

```python
import trafilatura, requests, json
from sentence_transformers import SentenceTransformer
import chromadb
```

```python
query = english_text
```

```python
import requests

n8n_webhook_url = "https://rrajdev.app.n8n.cloud/webhook/farm-query"
response = requests.post(n8n_webhook_url, json={"query": query})
response = requests.post(n8n_webhook_url, json={"query": query})

print("Status Code:", response.status_code)
print("Raw Response:", response.text)

try:
    print("JSON Response:", response.json())
except Exception as e:
    print("Error parsing JSON:", e)


# response should contain top URLs
urls = response.json()["urls"]
```

```python
urls
```

```python
import requests
import trafilatura

def fetch_and_extract(url):
    try:
        response = requests.get(url, timeout=60, headers={"User-Agent": "Mozilla/5.0"})
        if response.status_code == 200:
            return trafilatura.extract(response.text)
    except Exception as e:
        print(f"Error fetching {url}: {e}")
    return None

cleaned = []
for url in urls:
    content = fetch_and_extract(url)
    if content:
        cleaned.append({"url": url, "content": content})
```

```python
cleaned
```

```python
chroma_client = chromadb.PersistentClient(path="./chroma_store")
collection = chroma_client.get_or_create_collection("farming_knowledge")
embedder = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
for i, entry in enumerate(cleaned):
    embedding = embedder.encode(entry["content"]).tolist()
    collection.add(
        ids=[f"doc_{i}"],
        documents=[entry["content"]],
        embeddings=[embedding],
        metadatas=[{"source": entry["url"]}]
    )
```

```python
query_embedding = embedder.encode(query).tolist()
results = collection.query(query_embeddings=[query_embedding], n_results=3)
context = "\n".join(results["documents"][0])
print("Retrieved Context:\n", context)
```

```python
from groq import Groq

# Initialize the Groq client
client = Groq(api_key="Your_Groq_API_Key")

# Keep only first 2000 characters or first 2048 tokens (approx)
trimmed_context = context[:2000]

# Format the prompt for Mistral (Groq supports it)
response = client.chat.completions.create(
    model="meta-llama/llama-4-scout-17b-16e-instruct",  # Groq's fastest Mistral-based model
    messages=[
        {
            "role": "system",
            "content": "You are an agricultural expert. Use the context to answer the question."
        },
        {
            "role": "user",
            "content": f"Context: {trimmed_context}\nQuestion: {query}"
        }
    ],
    temperature=0.5,
    max_tokens=150
)

print(response.choices[0].message.content)
```

```python
from gtts import gTTS
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import IPython.display as display

# Step 1: Translate English → Hindi using NLLB-200
trans_model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
trans_tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")

def translate_to_hindi(english_text):
    translator = pipeline(
        "translation",
        model=trans_model,
        tokenizer=trans_tokenizer,
        src_lang="eng_Latn",
        tgt_lang="hin_Deva",
        max_length=400
    )
    result = translator(english_text)
    return result[0]['translation_text']


# Step 2: Convert Hindi Text → Audio
def hindi_tts(text, output_path="hindi_output.mp3"):
    tts = gTTS(text=text, lang='hi')
    tts.save(output_path)
    return output_path
```

```python
# Step 3: Run the full flow
english_input = response.choices[0].message.content
hindi_translation = translate_to_hindi(english_input)
print("Hindi Translation:", hindi_translation)
```

```python
audio_path = hindi_tts(hindi_translation)
display.Audio(audio_path)
```
